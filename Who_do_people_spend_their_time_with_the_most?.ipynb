```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/Coding Dojo/04 Week 4: Explanatory Data Viz/time-spent-with-relationships-by-age-us.xlsx'
timedf = pd.read_excel(filename)
timedf.head()
```

```python
timedf.info()
```

```python
timedf = timedf.rename(columns= {"Time spent alone, by age of respondent (United States)" : "Alone",
                       "Time spent with friends, by age of respondent (United States)" : "With Friends",
                       "Time spent with children, by age of respondent (United States)" : "With Children",
                       "Time spent with with parents, siblings and other family, by age of respondent (United States)" : "With Family Members",
                       "Time spent with partner, by age of respondent (United States)" : "With Partner",
                       "Time spent with coworkers, by age of respondent (United States)" : "With Coworkers"})
timedf.head()
```

```python
timedf = timedf.drop(columns = ["Code"])
timedf.head()
```

```python
fig, ax = plt.subplots(figsize = (12,8))

ax = sns.lineplot(data = timedf, x = "Year", y = "Alone", marker = "o")

ax = sns.lineplot(data = timedf, x = "Year", y = "With Friends", marker = "o")

ax = sns.lineplot(data = timedf, x = "Year", y = "With Children", marker = "o")

ax = sns.lineplot(data = timedf, x = "Year", y = "With Family Members", marker = "o")

ax = sns.lineplot(data = timedf, x = "Year", y = "With Partner", marker = "o")

ax = sns.lineplot(data = timedf, x = "Year", y = "With Coworkers", marker = "o")

ax.set_title("Who Americans Spend Their Time With by Age", fontsize = 20, fontweight = "bold")
ax.set_xlabel("Age", fontsize = 15)
ax.set_ylabel("Time Spent per Day (Minutes)", fontsize = 15);
ax.legend(loc = "best", labels = ["Alone", "With Friends", "With Children", "With Family Members", "With Partner", "With Coworkers"]);
```
